```python
cust = customer_data.copy()
cust["Age"] = 2026 - cust["Year_Birth"]
age = cust[["Year_Birth", "Age"]]
display(age.head(10))
```

```python
cust["Total_Purchases"] = (
    cust["NumWebPurchases"] +
    cust["NumCatalogPurchases"] +
    cust["NumStorePurchases"])
purchase = cust[["ID", "Total_Purchases"]]
display(purchase.head(10))
```

```python
cust["Total_Spent"] = (cust["MntWines"] + cust["MntFruits"] + cust["MntMeatProducts"] + cust["MntFishProducts"] + cust["MntSweetProducts"] + cust["MntGoldProds"])
```

```python
display(cust.columns)
```

```python
cust["Customer_Value"] = cust["Total_Spent"] * cust["Total_Purchases"]
value = cust[["ID", "Total_Spent", "Total_Purchases", "Customer_Value"]]
display(value.head(10))
```

```python
cust["Segment"] = "Low"

cust.loc[cust["Customer_Value"] >= 5000, "Segment"] = "Medium"
cust.loc[cust["Customer_Value"] >= 15000, "Segment"] = "High"
```

```python
segment = cust[["ID", "Customer_Value", "Segment"]]
display(segment.head(10))
```

```python
seg_count = cust["Segment"].value_counts()
display(seg_count)
```

```python
high = cust[cust["Segment"] == "High"]
display(high.head(10))
```

```python
medium = cust[cust["Segment"] == "Medium"]
display(medium.head(10))
```

```python
low = cust[cust["Segment"] == "Low"]
display(low.head(10))
```

```python
email = cust[["ID", "Income", "Age", "Total_Spent", "Total_Purchases", "Customer_Value", "Segment"]]
display(email.head(10))
```

```python
import matplotlib.pyplot as plt
seg = cust["Segment"].value_counts()
plt.figure(figsize=(6,4))
plt.bar(seg.index, seg.values)
plt.title("Customer Segment Distribution")
plt.xlabel("Segment")
plt.ylabel("Customers")
plt.show()
```

```python
age = cust["Age"]
plt.figure(figsize=(6,4))
plt.hist(age, bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Customers")
plt.show()
```

```python
inc = cust["Income"].dropna()
plt.figure(figsize=(6,4))
plt.hist(inc, bins=10)
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Customers")
plt.show()
```

```python
sp = cust.groupby("Segment")["Total_Spent"].mean()
plt.figure(figsize=(6,4))
plt.bar(sp.index, sp.values)
plt.title("Average Spending by Segment")
plt.xlabel("Segment")
plt.ylabel("Average Spending")
plt.show()
```

```python
res = cust["Response"].value_counts()
plt.figure(figsize=(6,4))
plt.bar(res.index.astype(str), res.values)
plt.title("Campaign Response")
plt.xlabel("Response")
plt.ylabel("Customers")
plt.show()
```
